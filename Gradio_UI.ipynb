```python
import gradio as gr
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import os
import matplotlib.pyplot as plt

# Load the model
model = load_model('cnn_lstm_model_optimized_reduced.h5')
#model = load_model('CNN_LSTM_AAA2.h5')

def process_video(video_path):
    # Create a temporary directory to store frames
    temp_dir = 'temp_frames'
    os.makedirs(temp_dir, exist_ok=True)

    # Open the video file
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    anomaly_scores = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Save the frame as an image
        frame_path = os.path.join(temp_dir, f'frame_{frame_count:04d}.jpg')
        cv2.imwrite(frame_path, frame)

        # Preprocess the frame (adjust as needed for your model)
        processed_frame = cv2.resize(frame, (224, 224))  # Adjust size as per your model's input
        processed_frame = processed_frame / 255.0  # Normalize

        # Make prediction
        prediction = model.predict(np.expand_dims(processed_frame, axis=0))
        anomaly_scores.append(prediction[0][0])

        frame_count += 1

    cap.release()

    # Clean up temporary files
    for file in os.listdir(temp_dir):
        os.remove(os.path.join(temp_dir, file))
    os.rmdir(temp_dir)

    # Create a plot of anomaly scores
    plt.figure(figsize=(10, 5))
    plt.plot(anomaly_scores)
    plt.title('Anomaly Scores')
    plt.xlabel('Frame')
    plt.ylabel('Anomaly Score')
    plot_path = 'anomaly_scores.png'
    plt.savefig(plot_path)
    plt.close()

    return video_path, plot_path

# Gradio interface
iface = gr.Interface(
    fn=process_video,
    inputs=gr.Video(),
    outputs=[
        gr.Video(label="Uploaded Video"),
        gr.Image(type="filepath", label="Anomaly Scores")
    ],
    title="Video Anomaly Detection",
    description="Upload a video to detect anomalies. The video will be displayed alongside the anomaly score plot."
)

iface.launch()
```
